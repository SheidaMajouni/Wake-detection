```python
from ultralytics import YOLO

# You're already using yolov8s.pt, which is a good step towards reducing capacity.
model = YOLO("yolov8s.pt")  # Using a smaller model (Small)

model.train(
    data="../Dataset/vessel_wakes.yaml",
    epochs=300,
    imgsz=640,
    project="runs-yoloS-tuned", # Changed project name to reflect tuning attempt
    name="train_reg_recall",
    # --- Regularization Parameters ---
    dropout=0.3,         # Add dropout for more robust learning
    weight_decay=0.0005, # Ensure weight decay is set, or slightly increased/decreased if needed
    # --- Data Augmentation (Crucial for Regularization) ---
    hsv_h=0.015,         # Adjust HSV Hue augmentation (default 0.015)
    hsv_s=0.7,           # Adjust HSV Saturation augmentation (default 0.7)
    hsv_v=0.4,           # Adjust HSV Value augmentation (default 0.4)
    degrees=5.0,         # Random rotation (default 0.0) - small rotation often helps
    translate=0.1,       # Random translation (default 0.1)
    scale=0.5,           # Random scaling (default 0.5)
    shear=0.0,           # Random shearing (default 0.0)
    perspective=0.0,     # Random perspective (default 0.0)
    flipud=0.0,          # Flip up-down (default 0.0) - maybe useful for wakes?
    fliplr=0.5,          # Flip left-right (default 0.5) - common and effective
    mosaic=0.5,          # Reduce Mosaic augmentation (default 1.0) if it's too aggressive
    mixup=0.0,           # Disable MixUp (default 0.0) - might not be beneficial for small objects
    copy_paste=0.0,      # Disable Copy-Paste (default 0.0)
    # --- Optimization Parameters for Convergence/Recall ---
    lr0=0.01,            # Initial learning rate (default 0.01)
    lrf=0.01,            # Final learning rate factor (default 0.01 - 1% of lr0)
    momentum=0.937,      # SGD momentum (default 0.937)
    # --- Early Stopping (VERY IMPORTANT for overfitting) ---
    patience=50          # Number of epochs to wait for val loss improvement before stopping
)

model.predict(
    source="../Dataset/images/valid",
    save=True,
    save_txt=True,
    conf=0.25,
    # For less FN, you might consider slightly lowering the confidence threshold for prediction.
    # However, this should primarily be tuned after the model is trained, not during.
    # If the model is not learning correctly, lowering this might just increase FPs.
    project="runs-yoloS-tuned",
    name="predict_reg_recall"
)
```
